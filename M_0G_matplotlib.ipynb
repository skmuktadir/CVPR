# Plotting graphs with Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
print(x)
print(y)
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y)
```

```python
plt.plot(x,y, color='green') 
```

```python
plt.plot(x,y, color='red', linewidth=7)
```

```python
plt.plot(x,y, color='green', linewidth=5, linestyle='dotted') 
```

```python
plt.plot(x,y,'b+--')
```

```python
plt.plot(x,y,color='blue',marker='+',linestyle='')
```

```python
plt.plot(x,y,color='blue',marker='+',linestyle='',markersize=20)
```

```python
plt.plot(x,y,'g<',alpha=0.5) # alpha can be specified on a scale 0 to 1
```

```python
import numpy as np
```

```python
# First create some toy data:
x = np.arange(-10, 10, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.figure()
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(['Sine','Cosine'])
plt.show()
```

```python
plt.figure()
plt.plot(x,y1,x,y2)
plt.legend(['Sine','Cosine'])
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.title('Sine')
plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title('Cosine')
plt.show()
```
