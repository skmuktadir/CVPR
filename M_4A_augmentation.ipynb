# Data Augmentation

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img

import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./resources/images/johnny_deep/explore.jpg')
#Converts from BGR to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.show()
```

```python
print(img.shape)

img_reshape = img.reshape(1,300,300,3)
print(img_reshape.shape)
```

```python
datagen = ImageDataGenerator(
    rotation_range=45,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.3,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest'
)
```

```python
i = 0
for batch in datagen.flow(img_reshape, 
                          save_to_dir='./resources/images/augmented/',
                          save_prefix='aug_',
                          save_format='jpg'
                         ):
    i += 1
    if i >= 20:
        break
```

```python
i = 0
for batch in datagen.flow_from_directory(directory='./resources/images/cat_dog/', 
                          batch_size=4,
                          save_to_dir='./resources/images/augmented/',
                          save_prefix='aug_',
                      save_format='jpg'
                         ):
    i += 1
    if i >= 2:
        break
```
